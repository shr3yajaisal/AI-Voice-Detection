## Human English Renaming

```python
folder_path = "/content/drive/MyDrive/IAPC Project/Dataset/Human Voices/Human English"

prefix = "human_eng"

files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.wav', '.mp3', '.flac'))]

files.sort()

for i, file in enumerate(files, start=1):
    ext = os.path.splitext(file)[1]
    new_name = f"{prefix}_{i:03d}{ext}"

    old_path = os.path.join(folder_path, file)
    new_path = os.path.join(folder_path, new_name)

    if not os.path.exists(new_path):
        os.rename(old_path, new_path)
    else:
        print(f"Skipped: {new_name}")

print("Renaming completed!")
```

```python
import pandas as pd
import librosa

base_path = "/content/drive/MyDrive/IAPC Project/Dataset"

data = []

folders = [
    ("AI", "English", os.path.join(base_path, "AI Voices", "AI English")),
    ("AI", "Hindi", os.path.join(base_path, "AI Voices", "AI Hindi")),
    ("Human", "English", os.path.join(base_path, "Human Voices", "Human English")),
    ("Human", "Hindi", os.path.join(base_path, "Human Voices", "Human Hindi"))
]

for label, language, folder_path in folders:

    if not os.path.exists(folder_path):
        print(f"Folder not found: {folder_path}")
        continue

    files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.wav', '.mp3', '.flac'))]

    for file in files:
        file_path = os.path.join(folder_path, file)

        try:
            duration = librosa.get_duration(path=file_path)

            data.append({
                "file_path": file_path,
                "label": label,
                "language": language,
                "duration": round(duration, 2)
            })

        except Exception as e:
            print(f"Error with file: {file_path}")
            continue

df = pd.DataFrame(data)

csv_path = os.path.join(base_path, "audio_dataset.csv")

df.to_csv(csv_path, index=False)

print(f"CSV created at: {csv_path}")
print(f"Total files processed: {len(df)}")
```
